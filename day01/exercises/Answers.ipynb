```python
import numpy as np
import time
import matplotlib.pyplot as plt
# Define the matrix dimensions
M = 100  # Rows
N = 100  # Columns
P = 100  # Number of columns for the second matrix

# Define the block size

# Create random matrices A and B
A = np.random.rand(M, N)
B = np.random.rand(N, P)
Time_list=[]
# Initialize the result matrix C
C = np.zeros((M, P))

# Perform matrix multiplication using blocks
for block_size in range(1,10):
  start=time.time()
  for i in range(0, M, block_size):
    for j in range(0, P, block_size):
        for k in range(0, N, block_size):
            # Extract blocks from matrices A and B
            A_block = A[i:i + block_size, k:k + block_size]
            B_block = B[k:k + block_size, j:j + block_size]

            # Perform block matrix multiplication
            C[i:i + block_size, j:j + block_size] += np.dot(A_block, B_block)
  end= time.time()
  Time_list.append(end-start)
# Print the result matrix C

plt.plot([1,2,3,4,5,6,7,8,9],Time_list,marker='o')
plt.show()
```
